```python


import requests

TOKEN = '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'.strip()  # Reemplázalo por tu token personal
HEADERS = {
    'Authorization': f'Bearer {TOKEN}',
    'Accept': 'application/vnd.github+json',
    'X-GitHub-Api-Version': '2022-11-28'
}

# 🔎 Buscar Repositorios Públicos

def search_repositories(query, per_page=5, page=1):
    url = 'https://api.github.com/search/repositories'
    params = {'q': query, 'per_page': per_page, 'page': page}
    response = requests.get(url, headers=HEADERS, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f'Error: {response.status_code} - {response.text}')
        return None

# Prueba
result = search_repositories('machine learning')
print(result)

# 📜 Obtener Commits y Contenido de Archivos

def list_commits(owner, repo, per_page=5, page=1):
    url = f'https://api.github.com/repos/{owner}/{repo}/commits'
    params = {'per_page': per_page, 'page': page}
    response = requests.get(url, headers=HEADERS, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f'Error: {response.status_code} - {response.text}')
        return None

def get_file_content(owner, repo, path):
    url = f'https://api.github.com/repos/{owner}/{repo}/contents/{path}'
    response = requests.get(url, headers=HEADERS)
    if response.status_code == 200:
        return response.json()
    else:
        print(f'Error: {response.status_code} - {response.text}')
        return None

# Ejemplos
commits = list_commits('psf', 'requests')
print(commits)

content = get_file_content('psf', 'requests', 'README.md')
print(content)

# 🔁 Paginación – Buscar varias páginas de repositorios

all_repos = []
for page in range(1, 4):  # 3 páginas
    data = search_repositories('machine learning', per_page=5, page=page)
    if data:
        all_repos.extend(data['items'])

print(f"Total repositorios obtenidos: {len(all_repos)}")

# 💾 Exportar resultados a CSV

import pandas as pd

df = pd.json_normalize(all_repos)
df.to_csv('repositorios_ml.csv', index=False)

df.head()
```
